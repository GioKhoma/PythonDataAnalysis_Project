```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

df = pd.read_csv('/content/sample_data/employee_satisfaction.csv')

print("მონაცემების გადახედვა:")
print(df.head())

scores = df['satisfaction_score']

mean = scores.mean()
median = scores.median()
mode_result = mode(scores, keepdims=True)
mode_value = mode_result.mode[0] if mode_result.count[0] > 0 else None
variance = scores.var()
std_dev = scores.std()
skewness = skew(scores)
kurt = kurtosis(scores)

print("\nსტატისტიკური მაჩვენებლები:")
print(f"საშუალო: {mean}")
print(f"მედიანა: {median}")
print(f"მოდა: {mode_value}")
print(f"დისპერსია: {variance}")
print(f"სტანდარტული გადახრა: {std_dev}")
print(f"ასიმეტრია: {skewness}")
print(f"ექსცესი: {kurt}")

plt.figure(figsize=(8, 6))
df['satisfaction_score'].plot(kind='hist', bins=10, title='კმაყოფილების ქულების ჰისტოგრამა', color='skyblue', edgecolor='black')
plt.xlabel('კმაყოფილების ქულა')
plt.ylabel('ხშირობა')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(8, 6))
df['satisfaction_score'].plot(kind='box', title='კმაყოფილების ქულების ბოქსპლოტი', vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'))
plt.xlabel('კმაყოფილების ქულა')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# შედეგების ინტერპრეტაცია
**სტატისტიკური მაჩვენებლები:**

მედიანა: გვაძლევს შუა მნიშვნელობას, რაც უფრო ზუსტია ასიმეტრიულ განაწილებებში.

მოდა: მიუთითებს ყველაზე გავრცელებულ ქულაზე.
დისპერსია და სტანდარტული გადახრა: მაღალი მაჩვენებლები აღნიშნავს, რომ მონაცემები უფრო ფართოდ არის განაწილებული.

ასიმეტრია: თუ ასიმეტრიის მაჩვენებელი დადებითია, მონაცემები მარჯვნივ არის გადახრილი.

ექსცესი: მაღალი ექსცესი ნიშნავს, რომ მონაცემებს აქვს "მძიმე კუდები" და უკიდურესი მნიშვნელობები.

ვიზუალიზაცია:
ჰისტოგრამა: მკაფიოდ აჩვენებს განაწილების ფორმას, არის თუ არა ის ნორმალური, ასიმეტრიული, ან არსებობს თუ არა უცნაური პიკური მნიშვნელობები.

ბოქსპლოტი: უფრო დეტალურად გამოყოფს უკიდურეს მნიშვნელობებს და საშუალებას გვაძლევს ვიზუალურად დავინახოთ განაწილების ასპექტები.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/sample_data/salaries.csv')

print("მონაცემების გადახედვა:")
print(df.head())

salaries = df['salary']

mean_salary = salaries.mean()
std_salary = salaries.std()

z_scores = (salaries - mean_salary) / std_salary

df['z_score'] = z_scores

outliers = df[np.abs(df['z_score']) > 3]

print("\nგამორჩეული მონაცემები (Z-Score მეთოდით):")
print(outliers)

outliers.to_csv('/content/sample_data/outliers_salaries.csv', index=False)

print("\nინტერპრეტაცია:")
print("გამორჩეული მონაცემები განისაზღვრება, როგორც ხელფასები, რომლებიც საშუალოდან 3 ან მეტ სტანდარტულ გადახრაზე შორს არიან.")
```

# დეტალური ახსნა
**1. მონაცემების ჩატვირთვა**

CSV ფაილი (salaries.csv) იტვირთება pandas-ის გამოყენებით, რომელიც საშუალებას გვაძლევს მონაცემებზე მარტივად ვიმუშაოთ.

**2. Z-Score-ების გამოთვლა**

Z-Score გამოითვლება ფორმულის მიხედვით:

  𝑧 = (𝑥−𝜇) / 𝜎
  
  სადაც:

  𝑥: ინდივიდუალური ხელფასი

  μ: ხელფასების საშუალო

  σ: სტანდარტული გადახრა

**3. გამორჩეული მონაცემების პოვნა**

Z-Score გვიჩვენებს, რამდენად შორს არის მონაცემები საშუალო მნიშვნელობიდან სტანდარტული გადახრის ერთეულებით. Z-Score-ით > 3 ან < -3 მონაცემები ითვლება გამორჩეულებად.

**4. შედეგების შენახვა**

გამორჩეული მონაცემები ინახება ახალ CSV ფაილში outliers_salaries.csv, რაც სასარგებლოა დეტალური ანალიზისთვის.

```python
import scipy.stats as stats

sample_size = 100
sample_defects = 8
claimed_defect_rate = 0.05
significance_level = 0.05

sample_proportion = sample_defects / sample_size
std_error = (claimed_defect_rate * (1 - claimed_defect_rate) / sample_size) ** 0.5
z_score = (sample_proportion - claimed_defect_rate) / std_error

p_value = stats.norm.cdf(z_score)

if p_value < significance_level:
    conclusion = "უარყოფილია ნულოვანი ჰიპოთეზა, მწარმოებლის მტკიცება მიღებულია."
else:
    conclusion = "ნულოვანი ჰიპოთეზის უარყოფა ვერ მოხერხდა, მწარმოებლის მტკიცება ვერ დადასტურდა."

print(f"z-ქულა: {z_score:.2f}")
print(f"p-მნიშვნელობა: {p_value:.4f}")
print(conclusion)
```

# დეტალური ახსნა და ინტერპრეტაცია
**მიდგომა**

  ჩვენი მიზანია, შევამოწმოთ, მართლაც არის თუ არა წუნდების მაჩვენებელი 5%-ზე ნაკლები. ამისთვის ვიყენებთ Z-ტესტს პროპორციებისთვის, რომელიც შედარებით მცირე შერჩევაზე გამართულ ტესტირებას საშუალებას გვაძლევს.


**მეთოდოლოგია**

  z-ქულის გამოთვლა:
  Z-ქულა გვიჩვენებს, რამდენად განსხვავდება შერჩევის პროპორცია მწარმოებლის მიერ დადგენილი წუნდების მაჩვენებლისგან სტანდარტული გადახრის ერთეულებით.
  ფორმულა:

  𝑧 = (𝑝−𝑃) / 𝑆

  სადაც:

  𝑝 – შერჩევის პროპორცია

  P – მოსალოდნელი წუნდების მაჩვენებელი (5%)

  SE – სტანდარტული შეცდომა

  p-მნიშვნელობის გამოთვლა:

  p-მნიშვნელობა გვიჩვენებს, რამდენად სავარაუდოა ასეთი შედეგის მიღება იმ შემთხვევაში, თუ ნულოვანი ჰიპოთეზა მართალია.

  დასკვნა:

  თუ p-მნიშვნელობა მნიშვნელოვნების დონეზე ნაკლებია (
  𝑝
  <
  0.05
  p<0.05), ნულოვანი ჰიპოთეზა უარყოფილია, და ჩვენ ვიღებთ მწარმოებლის მტკიცებას.


**ინტერპრეტაცია**

  თუ p-მნიშვნელობა ნაკლებია
  0.05-ზე, მაშინ შეგვიძლია ვთქვათ, რომ წუნდების მაჩვენებელი 5%-ზე ნაკლებია, და მწარმოებლის მტკიცება მიღებულია.
  თუ p-მნიშვნელობა
  0.05-ზე მეტია, ნულოვანი ჰიპოთეზის უარყოფა ვერ მოხერხდა და ვერ დავადასტურებთ, რომ წუნდების მაჩვენებელი 5%-ზე ნაკლებია.
